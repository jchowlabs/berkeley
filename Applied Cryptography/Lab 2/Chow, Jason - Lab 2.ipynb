```python
#Factoring of N0 without using built-in Python or SageMath functions. 

import time

def calculator(number):
    
    factors = []
    x = 2
    
    start = time.time()
    while(number > 1):
        if(number % x == 0):
            factors.append(x)
            number = number / x
        else:
            x = x + 1
    end = time.time()
    
    print("Seconds to factor: " + str(end-start))
    print("Factors: ", factors)

N0 = 2263
calculator(N0)
```

```python
#Using sqrts and CRT notation to calculate factors of N0. 

from itertools import combinations

#Calculates GCD given two values.
def calculate_gcd(a,b):
    
    if a < b:
        (a,b) = (b,a)

    if(a%b) == 0:
        return b
    else:
        return (calculate_gcd(b, a % b)) 

#Calculates the combinations, given number of items and elements.
def calculate_combinations(sqrts, elements):
  
    return list(combinations(sqrts, elements))

def main():
    
    N = 2263 #Value of N0
    sqrts = [924,1339,2238] #Sqrts calculated in Part 2
    combinations = (calculate_combinations(sqrts, 2)) #Calculates all 2 pair combinations of sqrts
    
    #Transforms list to individual pairs 
    combo1 = combinations[0] 
    combo2 = combinations[1]
    combo3 = combinations[2]
    
    #Sums each pair to be used as input to GCD algorithm
    sum_combo1 = sum(combo1) 
    sum_combo2 = sum(combo2)
    sum_combo3 = sum(combo3)

    #Calculates GCD by using N and summation of each sqrt combination (CRT equation)
    result1 = calculate_gcd(sum_combo1, N) 
    result2 = calculate_gcd(sum_combo2, N)
    result3 = calculate_gcd(sum_combo3, N)
    
    print("The sqrts of", N, "are: ", sqrts)
    print("The combinations of sqrts are: ", combinations)
    print("The GCD of (", sum_combo1, ",", N, ") is", result1)
    print("The GCD of (", sum_combo2, ",", N, ") is", result2)
    print("The GCD of (", sum_combo3, ",", N, ") is", result3)
    
main()
```
