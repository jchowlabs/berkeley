```python
import numpy as np
import pandas as pd
```

```python
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
creditcard_csv_path = '/content/drive/My Drive/MICS-CYBER-W207-Spring2022/For Students/Notebooks/creditcard.csv'
full_data = pd.read_csv(creditcard_csv_path)
```

```python
full_features = full_data.drop(["Time","Class"],axis=1)
full_labels = pd.DataFrame(full_data[["Class"]])
```

```python
full_features.describe()
```

```python
full_features_array = full_features.values
full_labels_array = full_labels.values
```

```python
train_features, test_features, train_labels, test_labels = \
    train_test_split(full_features_array, full_labels_array, train_size=0.90)
```

```python
standard_scaler = StandardScaler().fit(train_features)

scaled_train_features = standard_scaler.transform(train_features)
scaled_test_features = standard_scaler.transform(test_features)
```

```python
pd.Series(scaled_train_features.flatten()).describe()
```

```python
knn = KNeighborsClassifier(n_neighbors=5,algorithm="kd_tree",n_jobs=-1)
```

```python
# numpy.ravel() returns a contiguous flattened array
# e.g. [[1, 2, 3], [4, 5, 6]] => [1, 2, 3, 4, 5, 6]
knn.fit(scaled_train_features, train_labels.ravel())
```

```python
knn_predicted_test_labels = knn.predict(scaled_test_features)
```

```python
tn, fp, fn, tp = confusion_matrix(test_labels, knn_predicted_test_labels).ravel()
```

```python
knn_accuracy_score = accuracy_score(test_labels, knn_predicted_test_labels)
knn_precison_score = precision_score(test_labels, knn_predicted_test_labels)
knn_recall_score = recall_score(test_labels, knn_predicted_test_labels)
knn_f1_score = f1_score(test_labels, knn_predicted_test_labels)
```

```python
print("Confusion Matrix")
print("tn =", tn, "fp =", fp)
print("fn =", fn, "tp =", tp)
```

```python
print("Accuracy -->", knn_accuracy_score)
print("Precison -->", knn_precison_score)
print("Recall   -->", knn_recall_score)
print("F1       -->", knn_f1_score)
```
